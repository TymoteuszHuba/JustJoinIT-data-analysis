```python
import os
import ijson
import pandas as pd

# Funkcja do strumieniowego wczytywania danych z pliku JSON
def process_json_file(file_path, output_csv, columns, first_file):
    with open(file_path, 'r', encoding='utf-8') as f:
        rows = []  # Tymczasowe przechowywanie wierszy
        for item in ijson.items(f, 'item'):  # Iteracja po elementach JSON
            row = {key: item.get(key, None) for key in columns}  # Wyciąganie potrzebnych kolumn
            rows.append(row)

            # Zapisuj dane partiami co 1000 wierszy
            if len(rows) >= 1000:
                pd.DataFrame(rows).to_csv(output_csv, mode='w' if first_file else 'a', 
                                          header=first_file, index=False)
                rows = []  # Wyczyść bufor
                first_file = False

        # Zapisz pozostałe dane
        if rows:
            pd.DataFrame(rows).to_csv(output_csv, mode='w' if first_file else 'a', 
                                      header=first_file, index=False)

# Ścieżka do folderu z plikami JSON
folder_path = '/Users/tymoteuszhuba/Desktop/Python/my_projects/JustJoinIT_project/data'
output_csv = 'processed_data.csv'

# Kolumny, które chcesz zachować
columns = ['title', 'marker_icon', 'workplace_type', 'company_name', 'city', 'country_code', 'company_size', 'experience_level', 'remote', 'skills', 'employment_types', 'published_at']

# Przetwarzanie plików w folderze
first_file = True
for file_name in os.listdir(folder_path):
    if file_name.endswith('.json'):
        file_path = os.path.join(folder_path, file_name)
        process_json_file(file_path, output_csv, columns, first_file)
        first_file = False

print(f"Dane zapisane do {output_csv}")
```
